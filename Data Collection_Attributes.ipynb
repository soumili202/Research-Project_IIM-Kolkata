```python
#Function for API CALL
def get_details(long,lat):
    url = "https://zillow56.p.rapidapi.com/search_coordinates"

    querystring = {"long":long,"lat":lat,"d":"0.5"}

    headers = {
        "X-RapidAPI-Key": "4f116457a0mshb4068f571586f3dp1de01fjsn205367bf6b0c",
        "X-RapidAPI-Host": "zillow56.p.rapidapi.com"
          }

    response = requests.get(url, headers=headers, params=querystring)

    return(response)
    
```

```python
#Function to Parse Requests
def parse_requests(lo,la):
    return(get_details(lo,la).json())
```
